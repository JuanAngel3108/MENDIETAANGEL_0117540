```python
#%% Leer una lista de números ya ordenados de forma ascendente y verificar que 
#dicha lista está ordenada. Luego, leer un número e insertarlo en la lista en 
#la posición que le corresponde a dicho número.
while True:
    try:
        cant_lista = int(input('Dígame cuantos números tiene la lista: '))
#El usuario debe de digitar la longitud de la lista deseada
        if cant_lista < 1:
            print("¡Imposible!")
        else:
            lista = []
            for i in range(cant_lista):
                numero = int(input(f'Dígame el número {i+1} entero: '))
                lista += [numero]
            print()
            print("La lista creada es:", lista)
#Este ciclo le permite al usuario digitar la lista número por número
            x = sorted(lista)
#Este comando nos orden la lista de menor a mayor
            if lista != x:
                print()
                print('¡ERROR! Su lista no está ordenada de forma ascendente. '
                      'Debe de digitar nuevamente la lista. ')
            else:
                num = int(input('Digite un número cualquiera: '))
                lista.append(num)
                for i in range(len(lista)-1,0,-1):
                    for j in range(i):
                        if lista[j]>lista[j+1]:
                            cambio = lista[j]
                            lista[j] = lista[j+1]
                            lista[j+1] = cambio
#Este ciclio le indica al programa ordenar el número ingresado por el usuario
#en la posición correcta
                            
                print() 
                print('La lista nueva es: ',lista)
            break
    except ValueError:
        print()
        print('¡ERROR! Escriba un valor numérico entero.')
        
```
