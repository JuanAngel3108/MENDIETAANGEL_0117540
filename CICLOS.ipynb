```python
#%% #Sin utilizar el comando abs, escribir un programa que calcule e imprima el 
#valor absoluto de cualquier número (sea entero o decimal).

while True:
    try:
        x = float(input('Digite un número: '))
#El usuario debe digitar un número
        print()
        if x >= 0:
#Si x es positivo el resultado del valor absoluto será el mismo número
            print(f'El valor absoluto de {x} es {x}.')
        else:
            print(f'El valor absoluto de {x} es {-x}.')
#Por el contrario si es negativo se convertirá en positivo su valor absoluto
        break
    except ValueError:
        print()
        print('¡ERROR! Digite un valor numérico.')
        
```

```python
#%% Leer dos números enteros y determinar si la magnitud de la diferencia entre
#los dos es un número primo.
        
while True:
    try:
        x = int(input('Digite un número: '))
        y = int(input('Digite otro número: '))
#El usuario debe digitar dos números enteros
        print()
        num = abs(x-y)
        print(f'La magnitud de la diferencia entre {x} y {y} es {num}.')
#Estas líneas imprimen la magnitud (poitiva) entre los números
        print()
        if num==2 or num==3 or num==5 or num==7:
            print(f'{num} es un número primo.')
#Si esa magnitud es 2,3,5 o 7 será número primo directamente
            break
        elif num<2 or num%2==0 or num%3==0 or num%4==0 or num%5==0 or num%7==0:
            print(f'{num} no es un número primo.')
#Si el residuo de la magnitud sobre 2,3,4,5 o 7 es igual a cero, el número 
            #no será un número primo
            break
        else:
            print(f'{num} es un número primo.')
#Por el contrario si el número es divisible por 1 y por el mismo es par.
        break              
    except ValueError:
        print()
        print('¡ERROR! Digite un valor numérico entero.')
```

```python
#%% Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
#corresponden al centro y al radio de dos círculos. Lea un punto de 
#coordenadas (a,b). Determinar si (a,b) está contenido: a) dentro del 
#círculo 1; b) dentro del círculo 2; c) dentro de ambos círculos; d) fuera de 
#ambos círculos. Recordemos que el círculo es el lugar geométrico de los puntos
#del plano cuya distancia a otro punto fijo, llamado centro, es menor o igual 
#que una cantidad constante, llamada radio. 

while True:
    try:
        import math #eSsta linea nos permite importar la biblioteca math
        #para poder usarla más adelante.
        x1 = float(input('Digite la coordenada x del centro de la primer'
                         ' circunferencia: ' ))
        y1 = float(input('Escriba la coordenada y del centro de la primera' 
                         ' circunferencia: '))
#Estes líneas le permiten digitar las coordenadas del centro de las
#circunferencia.
        r1 = float(input('Digite el radio de la primer cincunferencia: '))
#Esta línea permite digitar el radio de la circunferencia
        print()
        print(f'La primera circunferencia tiene su centro en {(x1,y1)} y '
              f'presenta un radio de {r1}.')
        print()
        x2 = float(input('Ahora escriba la coordenada x del centro de la '
                         'segunda circunferencia: '))
        y2 = float(input('Esta vez digite la coordenada y del centro de la '
                         'segunda circunferencia: '))
         #Estes líneas le permiten digitar las coordenadas del centro de las
        #circunferencias.
        r2 = float(input('Escriba el radio de la segunda circunferencia: '))
        print()
        #Esta línea permite digitar el radio de la circunferencia
        print(f'La segunda circunferencia tiene su centro en {(x2,y2)} y '
              f'presenta un radio de {r2}.')
        print()
        a = float(input('Digite la coordenada x del punto: '))
        b = float(input('Escriba la coordenada y del punto: '))
        print()
   #Estas lineas le indican al usuario que debe digitar un puntos
#cualesquiera para ser analizados 
        
        d1 = math.sqrt((a-x1)**2+(b-y1)**2)
        d2 = math.sqrt((a-x2)**2+(b-y2)**2)
        #Estas lineas le indican al programa la distancia que hay entre el 
        #centro de las circunferencias y el punto digitado anteriormente
        #sabiendo que la distancia de un punto hacia los centros será la
        #raiz de la diferencia, elevada al cuadrado de coordenadas x y y
        #con sus homónimos en el punto.
        
        if d1 > r1 and d2 > r2:
            print(f'El punto {(a,b)} está afuera de ambas circunferencias.')
        elif d1 > r1 and d2 <= r2:
            print(f'El punto {(a,b)} está dentro solamente de la '
                  'circunferencia 2.')
        elif d1 <= r1 and d2 > r2:
            print(f'El punto {(a,b)} está dentro solamente de la '
                  'circunferencia 1.')
        else:
            print(f'El punto {(a,b)} esta dentro de ambas circunferencias.')
 #Estas línas le permiten saber al usuario donde se encuentra
        #el punto dado. Esto se logra gracias a los operadores lógicos 
        #(comparaciones), en este caso solo se usó and.
        break        
    except ValueError:
        print()
        print('¡ERROR! Digite un valor numérico.')
```
